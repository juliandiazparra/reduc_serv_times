```python
#Importo archivo de tickets
filepath=os.path.join(tickets,'Lista de Ticket Cerrados')
lista_tickets=pd.read_csv(filepath+'.csv')
lista_tickets.info()
```

```python
base_tickets=lista_tickets.iloc[:, [0, 1, 2, 3, 4, 5, 9, 12, 13]]
base_tickets.head()
```

```python
base_tickets['Asunto'].value_counts()
```

```python
#Creo categorias
#Tecnicos
base_tickets.loc[:,'instal']=base_tickets['Asunto'].str.contains('instal', 
                                                                   case=False)*1
base_tickets.loc[:,'soporte']=base_tickets['Asunto'].str.contains(
    'soporte|sin servicio|revisar|lento|intermiten|potencia|servicio tecnico|ocultar|pegar|alinear|configurar|clave]equipo|finalizar', 
                                                                    case=False)*1
base_tickets.loc[:,'traslado']=base_tickets['Asunto'].str.contains('trasla|transla', 
                                                                     case=False)*1
base_tickets.loc[:,'retiro']=base_tickets['Asunto'].str.contains('retir', 
                                                                   case=False)*1
base_tickets.loc[:,'mantenimiento']=base_tickets['Asunto'].str.contains('mantenimien', 
                                                                          case=False)*1

#Comercial
base_tickets.loc[:,'facturacion']=base_tickets['Asunto'].str.contains('facturacion|factura|editar', 
                                                                        case=False)*1
base_tickets.loc[:,'cambio_plan']=base_tickets['Asunto'].str.contains(r'^(?=.*cambio)(?=.*plan)', 
                                                                        case=False)*1
base_tickets.loc[:,'suspendido']=base_tickets['Asunto'].str.contains('suspend', case=False)*1

#Identificadores
base_tickets.loc[:,'tv']=base_tickets['Asunto'].str.contains('tv', 
                                                               case=False)*1
base_tickets.loc[:, 'llamada']=base_tickets['Asunto'].str.contains('llamada', 
                                                                     case=False)*1
```

```python
#Creo variable de categoria
#soporte
base_tickets.loc[(base_tickets['soporte']==1),'categoria']='Soporte'

#instal
base_tickets.loc[(base_tickets['instal']==1) & 
                  (base_tickets['soporte']==0),'categoria']='Instalacion'
#Retiro
base_tickets.loc[(base_tickets['retiro']==1) & 
                  (base_tickets['soporte']==0),'categoria']='Retiro'
#Traslado
base_tickets.loc[(base_tickets['traslado']==1), 'categoria']='Traslado'

#Cambio de plan
base_tickets.loc[(base_tickets['cambio_plan']==1), 'categoria']='Cambio de plan'
base_tickets.loc[(base_tickets['facturacion']==1), 'categoria']='Facturacion'
base_tickets.loc[(base_tickets['suspendido']==1), 'categoria']='Suspendido'
```

```python
#Miro formatos de fechas
lista_tickets[['Fecha', 'F. Cerrado']]
```

```python
#Doy formato fechas
#Fecha abierto
base_tickets.loc[:,'fecha_abierto'] = pd.to_datetime(
    base_tickets['Fecha'], format='%d/%m/%Y %I:%M:%S %p')
#Fecha cerrado
base_tickets.loc[:,'fecha_cerrado'] = pd.to_datetime( 
    base_tickets['F. Cerrado'], format='%d/%m/%Y %H:%M:%S')
```

```python
#Miro variables relevantes
base_tickets.columns
```

```python
#Depuro variables
tickets_df=base_tickets.loc[:, list(base_tickets.columns[np.r_[0:5, 6, 8:22]])]
tickets_df.columns
```

```python
#Renombro variables
tickets_df.columns=['id_ticket', 'departamento', 'nombre', 
 'asunto', 'tecnico', 'zona', 
 'abierto_por'] +list(base_tickets.columns[9:])
tickets_df.info()
```

```python
#Guardo la base como csv
years=list(tickets_df['fecha_cerrado'].dt.year.unique())
for i in years:
    file_path=os.path.join(tickets_clean, 'Tickets_'+str(i))
    mask_year=pd.to_datetime(tickets_df['fecha_cerrado']).dt.year==i
    tickets_df.loc[mask_year].to_excel(file_path+'.xlsx', index=False)
    tickets_df.loc[mask_year].to_csv(file_path+'.csv', index=False)
```
