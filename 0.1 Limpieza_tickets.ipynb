```python
base_tickets=lista_tickets.iloc[:, [0, 1, 2, 3, 4, 5, 9, 12, 13,14]]
base_tickets.head()
```

```python
base_tickets['Asunto'] = base_tickets['Asunto'].str.lower().str.strip()

# Diccionario de categorías técnicas y administrativas
categorias_tecnicas = {
    'instal': 'instal',
    'soporte': r"""soporte|sin servicio|revisar|lento|lentitud|intermiten|potencia
    |servicio.*tecnico|ocultar|pegar|alinear|configurar|finalizar|cable|repetid|equipo|router|certificar""",
    'traslado': 'trasla|transla',
    'retiro': 'retir',
    'mantenimiento': 'mantenimien',
    'migracion':'migracion'
}

categorias_administrativas = {
    'facturacion': 'facturacion|factura|editar',
    'cambio_plan': r'^(?=.*cambio)(?=.*plan)',
    'suspendido': 'suspend',
    'recoleccion':'recolecc',
    'cambio_contraseña': r'^(?=.*cambio)(?=.*(clave|contraseña))',
    'cambio_titular':r'^(?=.*cambio)(?=.*titular)'
}

# Diccionario de identificadores adicionales
identificadores = {
    'tv': 'tv',
    'internet': 'internet'
}

# Orden de prioridad para las categorías técnicas y administrativas
prioridad_tecnicas = ['soporte','traslado','instal', 'retiro', 'mantenimiento','migracion']
prioridad_administrativas = ['facturacion', 'cambio_plan', 'suspendido', 'recoleccion', 'cambio_contraseña','cambio_titular']

# Inicializamos la columna de categoría en None
base_tickets['categoria'] = None

# Clasificar primero las categorías administrativas
for categoria in prioridad_administrativas:
    base_tickets.loc[
        base_tickets['categoria'].isnull() & 
        base_tickets['Asunto'].str.contains(categorias_administrativas[categoria], case=False), 
        'categoria'
    ] = categoria

# Clasificar primero las categorías técnicas, asegurando que no se clasifiquen como soporte si ya están en recolección
for categoria in prioridad_tecnicas:
    base_tickets.loc[
        base_tickets['categoria'].isnull() & 
        base_tickets['Asunto'].str.contains(categorias_tecnicas[categoria], case=False) & 
        ~(base_tickets['Asunto'].str.contains('recoleccion', case=False)),
        'categoria'
    ] = categoria

# Crear una nueva columna 'departamento' para identificar si es técnico o administrativo
base_tickets['departamento'] = base_tickets['categoria'].apply(
    lambda cat:
    'tecnico' if cat in categorias_tecnicas else 
    'administrativo' if cat in categorias_administrativas else
    'indefinido'
)
```

Creo sub-categorias de soporte

```python
# Diccionario de subcategorías para soporte
subcategorias_soporte = {
    'rendimiento_servicio': r'lento|intermiten',
    'falta_de_servicio': r'sin servicio',
    'revisiones_tecnicas': r'revisar',
    'instalacion_configuracion': r'soporte instalar|certificar canal|cable'
}

# Inicializar la columna de subcategoría en None
base_tickets['subcategoria'] = None

# Clasificar subcategorías solo para tickets de soporte
for subcategoria, pattern in subcategorias_soporte.items():
    base_tickets.loc[
        (base_tickets['categoria'] == 'soporte') & 
        base_tickets['Asunto'].str.contains(pattern, case=False),
        'subcategoria'
    ] = subcategoria
```

```python
base_tickets.loc[base_tickets['categoria']=='soporte','subcategoria'].value_counts()
```

```python
# Filtrar los asuntos de la categoría soporte
asuntos_soporte = base_tickets[base_tickets['categoria'] == 'soporte']['Asunto'].tolist()

def preprocess_text(text):
    text = text.lower()  # Convertir a minúsculas
    text = re.sub(r'[^a-z\s]', '', text)  # Eliminar caracteres especiales
    return text.split()  # Tokenizar

# Preprocesar los asuntos
asuntos_soporte_preprocesados = [preprocess_text(asunto) for asunto in asuntos_soporte]
```

```python
from collections import Counter

stop_words = set([
    'servicio', 'soporte', 'de', 'del', 'la', 'el', 'y', 'en', 'con', 'por', 
    'a', 'que', 'los', 'las', 'un', 'una', 'esto', 'esto', 'por', 'como', 'más','cliente',
    'tecnico'
    # Puedes añadir más palabras según sea necesario
])
# Contar las palabras, excluyendo las stop words
filtered_word_counts = Counter(
    word for word in [word for asunto in asuntos_soporte_preprocesados for word in asunto]
    if word not in stop_words
)

# Imprimir las 10 palabras más comunes sin las stop words
print(filtered_word_counts.most_common(10))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans

stop_words = [
    'servicio', 'soporte', 'de', 'del', 'la', 'el', 'y', 'en', 'con', 'por', 
    'a', 'que', 'los', 'las', 'un', 'una', 'esto', 'esto', 'por', 'como', 'más','cliente',
    'tecnico'
    # Puedes añadir más palabras según sea necesario
]

# Crear una matriz de términos con stop words excluidas
vectorizer = CountVectorizer(stop_words=stop_words)
X = vectorizer.fit_transform([' '.join(asunto) for asunto in asuntos_soporte_preprocesados])

# Aplicar K-Means
k = 5  # Define el número de clusters que deseas
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)

# Agrupar los asuntos por el cluster
base_tickets.loc[base_tickets['categoria']=='soporte','cluster'] = kmeans.labels_
```

```python
for i in range(k):
    print(f"Cluster {i}:")
    print(base_tickets[base_tickets['cluster'] == i]['Asunto'].sample(5).tolist())
```
