```python
base_tickets=lista_tickets.iloc[:, [0, 1, 2, 3, 4, 5, 9, 12, 13,14]]
base_tickets.head()
```

```python
base_tickets['Asunto'] = base_tickets['Asunto'].str.lower().str.strip()

# Diccionario de categorías técnicas y administrativas
categorias_tecnicas = {
    'instal': 'instal',
    'soporte': r"soporte|sin servicio|revisar|lento|lentitud|intermiten|potencia|servicio.*tecnico|ocultar|pegar|alinear|configurar|finalizar|cable|repetid|equipo|router|certificar",
    'traslado': 'trasla|transla',
    'retiro': 'retir',
    'mantenimiento': 'mantenimien',
    'migracion':'migracion'
}

categorias_administrativas = {
    'facturacion': 'facturacion|factura|editar',
    'cambio_plan': r'^(?=.*cambio)(?=.*plan)',
    'suspendido': 'suspend',
    'recoleccion':'recolecc',
    'cambio_contraseña': r'^(?=.*cambio)(?=.*(clave|contraseña))',
    'cambio_titular':r'^(?=.*cambio)(?=.*titular)'
}

# Diccionario de identificadores adicionales
identificadores = {
    'tv': 'tv',
    'internet': 'internet'
}

# Orden de prioridad para las categorías técnicas y administrativas
prioridad_tecnicas = ['soporte','traslado','instal', 'retiro', 'mantenimiento','migracion']
prioridad_administrativas = ['facturacion', 'cambio_plan', 'suspendido', 'recoleccion', 'cambio_contraseña','cambio_titular']

# Inicializamos la columna de categoría en None
base_tickets['categoria'] = None

# Clasificar primero las categorías técnicas
for categoria in prioridad_tecnicas:
    base_tickets.loc[base_tickets['categoria'].isnull() & base_tickets['Asunto'].str.contains(categorias_tecnicas[categoria], case=False), 'categoria'] = categoria

# Clasificar las categorías administrativas solo si aún no tienen categoría técnica
for categoria in prioridad_administrativas:
    base_tickets.loc[base_tickets['categoria'].isnull() & base_tickets['Asunto'].str.contains(categorias_administrativas[categoria], case=False), 'categoria'] = categoria

# Crear una nueva columna 'departamento' para identificar si es técnico o administrativo
base_tickets['departamento'] = base_tickets['categoria'].apply(
    lambda cat:
    'tecnico' if cat in categorias_tecnicas else 
    'administrativo' if cat in categorias_administrativas else
    'indefinido'
)
```

```python
# Inicializar el diccionario para los resultados
departamentos_categorias_actualizados = {}

# Tickets del departamento técnico
departamentos_categorias_actualizados['tecnico'] = {
    categoria: base_tickets[base_tickets['categoria'] == categoria]['Asunto'].unique().tolist()
    for categoria in categorias_tecnicas.keys()
}

# Tickets del departamento administrativo
departamentos_categorias_actualizados['administrativo'] = {
    categoria: base_tickets[base_tickets['categoria'] == categoria]['Asunto'].unique().tolist()
    for categoria in categorias_administrativas.keys()
}

# Verificar los resultados
for departamento, categorias in departamentos_categorias_actualizados.items():
    print(f"\nDepartamento: {departamento}")
    for categoria, asuntos in categorias.items():
        print(f"\nCategoría: {categoria}")
        for asunto in asuntos:
            print(asunto)
```
